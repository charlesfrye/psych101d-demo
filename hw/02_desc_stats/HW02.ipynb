## 1. Proving Useful Facts About Means

$$\bar x = \frac{1}{n}\sum_{i=1}^n x_i$$

**a)** Show that if all elements of a list are the same value, then the average of that list is equal to said common value.

$$ \bar x = \frac{1}{n} \sum_{i=1}^n x_i $$
$$ x_i = c $$
$$ = \frac{1}{n} \sum_{i=1}^n c $$
$$ = \frac{cn}{n} $$
$$ = c $$

**b)** Show that if every element of a list is increased by a number, the average also increases by said number.

$$ \frac{1}{n} \sum_{i=1}^n x_i + c $$
$$ \frac{1}{n} \sum_{i=1}^n x_i + \frac{1}{n}\sum_{i=1}^n c $$
$$ \bar x + \frac{cn}{n} $$
$$ \bar x + c $$

**c)** Show that if every element of a list is multiplied by a number, the average also is also multiplied by said number.

$$ \frac{1}{n} \sum_{i=1}^n cx_i $$
$$ \frac{c}{n} \sum_{i=1}^n x_i$$
$$ \frac{\bar x  c n}{n}$$
$$ c \bar x $$
